```python
import numpy as np
import networkx as nx
import os
from random import shuffle          # importamos las librerias necesarias
import matplotlib.pylab as plt       
%matplotlib inline
from matplotlib.pyplot import title,xlabel,ylabel,show
```

```python
def ldata(archive):
    f=open(archive)
    data=[]
    for line in f:
        line=line.strip()
        col=line.split()
        data.append(col)
    return data
```

```python
def grafo(datosRed):
    G = nx.Graph()
    for i in range(np.shape(datosRed)[0]):
        G.add_edges_from([(datosRed[i][0],datosRed[i][1])])
    return G 
```

```python
APMS = ldata('yeast_AP-MS.txt')
grafoAPMS = grafo(APMS)

LIT = ldata('yeast_LIT.txt')
grafoLIT = grafo(LIT)


#LITREG = ldata('yeast_LIT_Reguly.txt')
#grafoLITREG = grafo(LITREG)

#Y2H = ldata('yeast_Y2H.txt')
#grafoY2H = grafo(Y2H)
```
